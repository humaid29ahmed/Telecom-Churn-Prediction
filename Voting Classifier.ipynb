# Voting Classifier

```python
import pandas as p
import numpy as n
import matplotlib.pyplot as plt
import seaborn as s
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=p.read_csv('churn.csv')
```

```python
data.columns
```

```python
del data['Customer ID']
del data['Gender']
del data['Zip Code']
del data['Latitude']
del data['Longitude']
del data['Total Charges']
del data['Total Revenue']
```

```python
df=data.dropna()
```

```python
data['Customer Status'].value_counts()
```

```python
df.drop(df.index[df['Customer Status'] == 'Joined'], inplace=True)
```

```python
df['Customer Status'].value_counts()
```

```python
df = df.rename({'Tenure in Months': 'Tenure_in_Months', 'Internet Service': 'Internet_Service',
                'Internet Type': 'Internet_Type', 'Avg Monthly GB Download': 'Avg_Monthly_GB_Download',
                'Unlimited Data': 'Unlimited_Data', 'Payment Method': 'Payment_Method',
                'Monthly Charge': 'Monthly_Charge', 'Total Charges': 'Total_Charges',
                'Total Extra Data Charges': 'Total_Extra_Data_Charges', 'Customer Status': 'Customer_Status'}, axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
'Offer', 'Internet_Type', 'Unlimited_Data', 'Contract', 'Payment_Method', 'Customer_Status'
var = []

for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
df
```

```python
#preprocessing, split test and dataset, split response variable
X = df.drop(labels='Customer_Status', axis=1)
#Response variable
y = df.loc[:,'Customer_Status']
```

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

ros =RandomOverSampler(random_state=42)
x_ros,y_ros=ros.fit_resample(X,y)
print("OUR DATASET COUNT         : ", Counter(y))
print("OVER SAMPLING DATA COUNT  : ", Counter(y_ros))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_ros, y_ros, test_size=0.30, random_state=1, stratify=y_ros)
print("Number of training dataset : ", len(X_train))
print("Number of test dataset     : ", len(X_test))
print("Total number of dataset    : ", len(X_train)+len(X_test))
```

Voting Classifier

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
svm = SVC()
dt= DecisionTreeClassifier()
lr = LogisticRegression()
rf = RandomForestClassifier()
```

```python
vot_clf = VotingClassifier(estimators=[('SVM', svm),('DecisionTree', dt), ('LogisticRegression', lr), ('RandomForest', rf)], voting='hard')
```

Training Process

```python
vot_clf.fit(X_train, y_train)
pred_vtng = vot_clf.predict(X_test)
```

Getting Accuracy

```python
accuracy = accuracy_score(y_test,pred_vtng)
print('Accuracy of Voting Classifier is: ',accuracy*100)
```

Finding Classification Report

```python
cr = classification_report(y_test,pred_vtng)
print('Classification report\n---------------------\n',cr)
```

Finding Confusion Matrix

```python
cm = confusion_matrix(y_test,pred_vtng)
print('Confusion matrix\n----------------\n',cm)
```

```python
DF = p.DataFrame()
DF["y_test"] = y_test
DF["predicted"] = pred_vtng
DF.reset_index(inplace=True)
plt.figure(figsize=(20, 5))
plt.plot(DF["predicted"][:100], marker='x', linestyle='dashed', color='red')
plt.plot(DF["y_test"][:100],  marker='o', linestyle='dashed', color='green')
plt.show()
```

```python
conf_matrix = confusion_matrix(y_test, pred_vtng)
#
# Print the confusion matrix using Matplotlib
#
fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```
